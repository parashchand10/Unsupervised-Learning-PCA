## **PCA - Principal Component Analysis**

PCA (Principal Component Analysis) is an unsupervised learning technique used for dimensionality reduction. It reduces the number of features (columns) while preserving most of the important information.

**Steps of PCA**:

1. Collect the Data.

2. Standardize the Data:
- Convert all features to the same scale.

3. Calculate the Covariance Matrix:
- Find how features are related to each other.

4. Calculate Eigenvalues and Eigenvectors:
- Eigenvalues tell us how much information (variance) each principal component contains.
- Eigenvectors determine the direction of the principal components.

5. Select Principal Components:
- Choose the components with the highest explained variance.

6. Transform the Data:
- Convert the original features into the new principal components.

```python
# import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples = 700, n_features = 5, centers = 3, cluster_std = 1, random_state = 42 )
```

```python
print("Original Data:", X.shape)
```

```python
# standardize feature
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# fit PCA
from sklearn.decomposition import PCA

# define the model
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X_scaled)
```

```python
print(f"After reduce data: {X_pca}")
```

```python
df = pd.DataFrame(X_pca, columns = ["PC1", "PC2"])
```

```python
df["label"] = y
```

```python
# scatter plot - visualize the graph correlation
sns.scatterplot(x = "PC1", y = "PC2", hue = "label", data = df, palette='Set2')
plt.show()
```

## **Conclusion**
1. Principal Component Analysis (PCA) was successfully applied to reduce the dimensionality of the dataset from 5 features to 2 principal components (PC1 and PC2).

2. The data was standardized using StandardScaler to ensure all features contributed equally.

3. The transformed data retained most of the important information while reducing complexity.

4. The scatter plot of PC1 and PC2 clearly shows the three clusters, demonstrating that PCA preserves the underlying structure of the data.
5. Overall, PCA reduced the number of features, improved data visualization, and prepared the dataset for faster and more efficient machine learning.
